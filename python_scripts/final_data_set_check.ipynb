```python
# Find duplicated rows based on 5-column uniqueness
duplicate_combo = final_dataset.duplicated(subset=['FIPS', 'State', 'County', 'County_file7', 'State_file7'], keep=False)

# Filter and display all such duplicated rows
duplicated_rows = final_dataset[duplicate_combo]

# Show the key columns to inspect
print(duplicated_rows[['FIPS', 'State', 'County', 'County_file7', 'State_file7']])
```

```python
# Step 1: Find rows where the 5-column combo is duplicated
duplicate_mask = final_dataset.duplicated(subset=['FIPS', 'State', 'County', 'County_file7', 'State_file7'], keep=False)

# Step 2: Filter those rows
duplicate_rows = final_dataset[duplicate_mask]

# Step 3: Group by those 5 columns to find exact duplicate groups
exact_duplicates = duplicate_rows.groupby(['FIPS', 'State', 'County', 'County_file7', 'State_file7']).filter(lambda x: len(x) > 1)

# Step 4: Display the exact duplicates (you can also save to Excel)
print(exact_duplicates[['FIPS', 'State', 'County', 'County_file7', 'State_file7']])
```

```python
# Step 1: Filter only rows for State 'Ak'
ak_data = final_dataset[final_dataset['State'] == 'Ak']

# Step 2: Find duplicates based on 5-column combination
duplicate_mask = ak_data.duplicated(subset=['FIPS', 'State', 'County', 'County_file7', 'State_file7'], keep=False)
ak_duplicates = ak_data[duplicate_mask]

# Step 3: Group and filter to show only those with true duplicates
ak_exact_duplicates = ak_duplicates.groupby(
    ['FIPS', 'State', 'County', 'County_file7', 'State_file7']
).filter(lambda x: len(x) > 1)

# Step 4: Display the result
print(ak_exact_duplicates[['FIPS', 'State', 'County', 'County_file7', 'State_file7']])
```
